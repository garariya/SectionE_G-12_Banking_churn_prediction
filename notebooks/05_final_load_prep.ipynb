# Final Data Load Preparation & KPI Computation

## Objective

Prepare a dashboard-ready dataset and compute key business metrics for reporting and Tableau visualization.

```python
import pandas as pd
import numpy as np
```

## Load Cleaned Dataset

```python
df = pd.read_csv("../data/processed/cleaned_banking_churn.csv")
df.head()
```

# KPI Computation

```python
total_customers = len(df)
churned_customers = df["churn"].sum()
retained_customers = total_customers - churned_customers
churn_rate = round(df["churn"].mean() * 100, 2)

avg_balance = round(df["current_balance"].mean(), 2)
avg_tenure = round(df["customer_tenure_years"].mean(), 2)

avg_credit = round(df["current_month_credit"].mean(), 2)
avg_debit = round(df["current_month_debit"].mean(), 2)
```

```python
kpi_summary = pd.DataFrame({
    "Metric": [
        "Total Customers",
        "Churned Customers",
        "Retained Customers",
        "Churn Rate %",
        "Average Balance",
        "Average Tenure (Years)",
        "Average Monthly Credit",
        "Average Monthly Debit"
    ],
    "Value": [
        total_customers,
        churned_customers,
        retained_customers,
        churn_rate,
        avg_balance,
        avg_tenure,
        avg_credit,
        avg_debit
    ]
})

kpi_summary
```

# Create Business-Friendly Columns

```python
df["churn_label"] = df["churn"].map({0:"Retained", 1:"Churned"})
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,25,35,45,55,65,100],
    labels=["<=25","26-35","36-45","46-55","56-65","65+"]
)
```

```python
df["balance_segment"] = pd.qcut(
    df["current_balance"],
    q=4,
    labels=["Low","Medium","High","Premium"]
)
```

# Final Validation

```python
df.shape
df.isnull().sum()
```

# Export Final Dashboard Dataset

```python
df.to_csv("../data/processed/final_dashboard_dataset.csv", index=False)
```

```python
kpi_summary.to_csv("../reports/kpi_summary.csv", index=False)
```
